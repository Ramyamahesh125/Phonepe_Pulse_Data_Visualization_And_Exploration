```python
import os
import json
import pandas as pd
import mysql.connector
```

```python
# Aggregated Insurance Data :

path1 = "C:/GUVI_Projects/pulse/data/aggregated/insurance/country/india/state/"
agg_insur_list = os.listdir(path1)

columns1 = {"States":[], "Years":[], "Quarters":[], "Transaction_Type":[], "Transaction_Count":[], "Transaction_Amount":[]}

for state in agg_insur_list :
    cur_state = path1 + state + "/"
    agg_insur_state = os.listdir(cur_state)

    for year in agg_insur_state:
        cur_year = cur_state + year + "/"
        agg_insur_year = os.listdir(cur_year)

        for file in agg_insur_year:
            cur_file = cur_year + file 
            data = open(cur_file, 'r')

            A = json.load(data)

            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                columns1['States'].append(state)
                columns1['Years'].append(year)
                columns1['Quarters'].append(int(file.strip(".json")))
                columns1['Transaction_Type'].append(name)
                columns1['Transaction_Count'].append(count)
                columns1['Transaction_Amount'].append(amount)

# Dataframe of Aggregated Insurance:

agg_insur_df = pd.DataFrame(columns1)


agg_insur_df['States']= agg_insur_df['States'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_insur_df['States']= agg_insur_df['States'].str.replace("-"," ")
agg_insur_df['States']= agg_insur_df['States'].str.replace('dadra & nagar haveli & daman & diu','dadra and nagar haveli and daman and diu')
agg_insur_df['States']= agg_insur_df['States'].str.title()

agg_insur_df
```

```python
# Aggregated Transacrion Data:

path2 =  "C:/GUVI_Projects/pulse/data/aggregated/transaction/country/india/state/"
agg_tran_list = os.listdir(path2)

columns2={"States":[], "Years":[], "Quarters":[], "Transaction_Type":[], "Transaction_Count":[], "Transaction_Amount":[]}

for state in agg_tran_list:
    cur_state = path2 + state + "/"
    agg_tran_state = os.listdir(cur_state)

    for year in agg_tran_state:
        cur_year = cur_state + year + "/"
        agg_tran_year = os.listdir(cur_year)

        for file in agg_tran_year:
            cur_file = cur_year + file
            data = open(cur_file,"r")

            B = json.load(data)
            
            for i in B['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                columns2['States'].append(state)
                columns2['Years'].append(year)
                columns2['Quarters'].append(int(file.strip('.json')))
                columns2['Transaction_Type'].append(name)
                columns2['Transaction_Count'].append(count)
                columns2['Transaction_Amount'].append(amount)
                
agg_tran_df = pd.DataFrame(columns2)

agg_tran_df['States']= agg_tran_df['States'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_tran_df['States']= agg_tran_df['States'].str.replace("-"," ")
agg_tran_df['States']= agg_tran_df['States'].str.replace('dadra & nagar haveli & daman & diu','dadra and nagar haveli and daman and diu')
agg_tran_df['States']= agg_tran_df['States'].str.title()

agg_tran_df
```

```python
# Aggregated User Data :

path3 = "C:/GUVI_Projects/pulse/data/aggregated/user/country/india/state/"
agg_user_list = os.listdir(path3)

columns3 ={"States":[], "Years":[], "Quarters":[], "Brands":[], "Transaction_Count":[], "Percentage":[]}

for state in agg_user_list :
    cur_state = path3 + state + '/'
    agg_user_state = os.listdir(cur_state)

    for year in agg_user_state:
        cur_year = cur_state + year + '/'
        agg_user_year = os.listdir(cur_year)

        for file in agg_user_year:
            cur_file = cur_year + file
            data = open(cur_file,'r')

            C = json.load(data)

            try:
                for i in C['data']['usersByDevice']:
                    brand = i['brand']
                    count = i['count']
                    percentage =i['percentage']
                    columns3['Brands'].append(brand)
                    columns3['Transaction_Count'].append(count)
                    columns3['Percentage'].append(percentage)
                    columns3['States'].append(state)
                    columns3['Years'].append(year)
                    columns3['Quarters'].append(int(file.strip('.json')))
                    
            except:
                pass

agg_user_df= pd.DataFrame(columns3)

agg_user_df['States']= agg_user_df['States'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
agg_user_df['States']= agg_user_df['States'].str.replace("-"," ")
agg_user_df['States']= agg_user_df['States'].str.replace('dadra & nagar haveli & daman & diu','dadra and nagar haveli and daman and diu')
agg_user_df['States']= agg_user_df['States'].str.title()
agg_user_df
```
